```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

### Task#1: Load/read images cat.jpeg and cameraman.tif [1 mark]

```python
# Load/read images cat.jpeg and cameraman.tif [1 mark]

img1 = cv2.imread('cat.jpeg')
img2 = cv2.imread('cameraman.tif', cv2.IMREAD_GRAYSCALE)
```

### Task#2: Add the two images, and display the result [2 marks]

```python
# Add the two images, and display the result [2 marks]
arr1 = np.array(img1, dtype=np.float32)
arr2 = np.array(img2, dtype=np.float32)

arr2_resized = cv2.resize(arr2, (arr1.shape[1], arr1.shape[0]))

added_image = arr1 + arr2_resized[:, :, np.newaxis]

plt.imshow(cv2.cvtColor(added_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Added Image')
plt.axis('off')
plt.show()
```

### Task#3: Using kernel1 and kernel2 apply convolution to cameraman.tif, and display the output image [2 marks]

```python
#Using kernel1 and kernel2 apply convolution to cameraman.tif, and display the output image [2 marks]
kernel1 = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])

kernel2 = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])


conv1 = cv2.filter2D(img2, -1, kernel1)
conv2 = cv2.filter2D(img2, -1, kernel2)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(conv1, cmap='gray')
axes[0].set_title('Convolution with Kernel1')
axes[0].axis('off')

axes[1].imshow(conv2, cmap='gray')
axes[1].set_title('Convolution with Kernel2')
axes[1].axis('off')

plt.show()
```
