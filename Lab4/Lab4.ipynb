## Task 1

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, exposure

img = data.moon()

p3, p80 = np.percentile(img, (3, 80))

img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

ax1.imshow(img_rescale, cmap='gray')
ax1.set_title('Rescaled Intensity (3rd - 80th Percentile)')
ax1.axis('off')

ax2.hist(img_rescale.flat, bins=256, range=(0, 255))
ax2.set_title('Histogram of Rescaled Image')
ax2.set_xlabel('Gray level values')
ax2.set_ylabel('Number of pixels')

plt.tight_layout()
plt.show()
```

## Task 2

```python
img_eq = exposure.equalize_hist(img)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

ax1.imshow(img_eq, cmap='gray')
ax1.set_title('Histogram Equalization (Moon)')
ax1.axis('off')

ax2.hist(img_eq.flat, bins=256)
ax2.set_title('Flattened Histogram')
ax2.set_xlabel('Normalized Intensity')
ax2.set_ylabel('Number of pixels')

plt.tight_layout()
plt.show()
```

## Task 3

```python
from skimage.exposure import match_histograms

reference = data.rocket()
image = data.chelsea()

matched = match_histograms(image, reference, channel_axis=-1)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.imshow(image)
ax1.set_title('Source (Chelsea)')
ax1.axis('off')

ax2.imshow(reference)
ax2.set_title('Reference (Rocket)')
ax2.axis('off')

ax3.imshow(matched)
ax3.set_title('Matched Image')
ax3.axis('off')

plt.tight_layout()
plt.show()
```
