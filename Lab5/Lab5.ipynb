## Task 1:


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/Parrot.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

smoothed_img = cv2.blur(img, (7, 7))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1), plt.imshow(img), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(smoothed_img), plt.title('7x7 Box Filtered'), plt.axis('off')
plt.show()
```

## Task 2:


```python

gaussian_5 = cv2.GaussianBlur(img, (5, 5), 0)

gaussian_21 = cv2.GaussianBlur(img, (21, 21), 0)

fig, ax = plt.subplots(1, 3, figsize=(18, 6))
ax[0].imshow(img), ax[0].set_title("Original"), ax[0].axis("off")
ax[1].imshow(gaussian_5), ax[1].set_title("5x5 Gaussian"), ax[1].axis("off")
ax[2].imshow(gaussian_21), ax[2].set_title("21x21 Gaussian"), ax[2].axis("off")
plt.show()
```

## Task 3:

```python
img_lap = cv2.imread('images/Parrot.jpg')
img_lap = cv2.cvtColor(img_lap, cv2.COLOR_BGR2RGB)

image_float = img_lap.astype(np.float32) / 255.0

laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

sharpened_lap = np.clip(image_float - laplacian, 0, 1)

fig, ax = plt.subplots(1, 3, figsize=(18, 6))
ax[0].imshow(image_float), ax[0].set_title("Original Image"), ax[0].axis("off")
ax[1].imshow(np.clip(laplacian, 0, 1)), ax[1].set_title("Laplacian Edge Map"), ax[1].axis("off")
ax[2].imshow(sharpened_lap), ax[2].set_title("Laplacian Sharpened"), ax[2].axis("off")
plt.show()
```
